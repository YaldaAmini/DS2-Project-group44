```python
def make_tweet_df(api, users=list, num_status=20, path='data/'):
    """
    """
    json_vars = ['created_at', 'id', 'text', 'entities', 'source',
                 'in_reply_to_status_id', 'user', 'geo', 'place',
                 'retweet_count', 'favorite_count', 'favorited', 'lang']
    
    sub_tags = ['entities', 'user', 'place']
    
    entities_tag = ['hashtags', 'urls']
    
    user_tag = ['id', 'name', 'screen_name', 'location', 'url',
                'protected', 'followers_count', 'friends_count',
                'created_at', 'geo_enabled', 'verified', 'lang']
    
    place_tag = ['id', 'place_type', 'name', 'country_code']
    
    col_names = ['created_at', 'id', 'text', 'hashtags', 'urls', 'source',
                 'in_reply_to_status_id', 'usr_id', 'usr_name', 'screen_name',
                 'usr_location', 'usr_url', 'protected', 'followers_count', 
                 'friends_count', 'usr_created', 'geo_enabled', 'verified',
                 'usr_lang', 'geo','id', 'place_type', 'name', 'country_code',
                 'retweet_count', 'favorite_count', 'favorited', 'lang']
    
    data = []
    users = list(users)
    
    for u in users:
        
        try:
            stats = api.user_timeline(u, count=num_status)
        except tweepy.TweepError:
            print('Error occured with users: {:s}, ignoring...'.format(u))
#             print(tweepy.TweepError.response.text)
        
        for s in stats:
            data_row = []
            
            for j in json_vars:
                if j in sub_tags:
                    if j == sub_tags[0]:
                        
                        for e in entities_tag:
                            if e==entities_tag[0]:
                                data_row.append(get_hashtags(s._json[j][e]))
                            else:
                                data_row.append(get_urls(s._json[j][e]))
                        
                    elif j == sub_tags[1]:
                        
                        for usr in user_tag:
                            data_row.append(s._json[j][usr])
                        
                    else:
                        if s._json[j]:
                            for p in place_tag:
                                data_row.append(s._json[j][p])
                        else:
                            for p in place_tag:
                                data_row.append(np.nan)
                    
                else:
                    data_row.append(s._json[j])
            
            data.append(data_row)
            
    df = pd.DataFrame(data, columns=col_names)
    
    final_num_ppl = len(df.usr_name.unique())
    num_tweets = len(df)
    
    print('Finished, dataframe with {:d} tweets from {:d} users created'.format(num_tweets,final_num_ppl))
    
    return df
```

```python
def get_hashtags(tag_dict):
    tags = []
    for i in range(len(tag_dict)):
        tags.append(tag_dict[i]['text'])
    return tags

def get_urls(url_dict):
    urls = []
    for i in range(len(url_dict)):
        urls.append(url_dict[i]['expanded_url'])
    return urls
```

```python
top_100 = pd.read_csv('top_100.csv')
top_100.head()
```

```python
api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
```

```python
df = make_tweet_df(api=api, users=top_100.id)
df.head()
```
